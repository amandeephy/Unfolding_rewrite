```python
import ROOT
import uproot
import numpy as np
import matplotlib.pyplot as plt
```

```python
BASE_DIR = '/scratch/brown/bakshi3/TopSpinCorr_Run2_generalized_ND/CMSSW_10_6_30/src/TopAnalysis/Configuration/analysis/diLeptonic/TUnfoldResults_2016/Nominal/combined/'
```

```python
uproot_fileptr = uproot.open(BASE_DIR + 'c_han.root')
```

```python
[k for k in uproot_fileptr.keys() if 'Theory' in k]
```

```python
unfolded = uproot_fileptr['c_hanTUnfResultCor_rebinnedA'].to_numpy()[0]
uproot_fileptr['c_hanTUnfResultCor_rebinnedA'].to_hist()
```

```python
theory_xsec = uproot_fileptr['c_hanTheoryXsec'].to_numpy()[0]
uproot_fileptr['c_hanTheoryXsec'].to_hist()
```

```python
unfolded  = np.divide(unfolded, np.sum(unfolded))
unfolded /= 0.33

theory_xsec  = np.divide(theory_xsec, np.sum(theory_xsec))
theory_xsec /= 0.33
```

```python
plt.plot(unfolded   , '-o', label = 'Unfolded Data')
plt.plot(theory_xsec, '-o', label = 'Theory xsec')
plt.ylim(0.4, 0.6)
plt.ylabel('AU')
plt.xlabel('Bins')
plt.title('Unfolded vs Theory c_han')
plt.legend()
```

```python
plt.plot(np.divide(unfolded, theory_xsec), '-o')
plt.ylim(0.95, 1.05)
plt.ylabel('Data/Theory')
plt.xlabel('Bins')
plt.title('Ratio c_han')
```

```python
varname = 'c_rkP'
uproot_fileptr = uproot.open(BASE_DIR +  varname + '.root')

unfolded    = uproot_fileptr[varname + 'TUnfResultCor_rebinnedA'].to_numpy()[0]
theory_xsec = uproot_fileptr[varname + 'TheoryXsec'].to_numpy()[0]

# Normalize and divide by bin width
unfolded  = np.divide(unfolded, np.sum(unfolded))
unfolded /= 0.33

theory_xsec  = np.divide(theory_xsec, np.sum(theory_xsec))
theory_xsec /= 0.33

plt.plot(unfolded   , '-o', label = 'Unfolded Data')
plt.plot(theory_xsec, '-o', label = 'Theory xsec')
plt.ylim(0.4, 0.6)
plt.ylabel('AU')
plt.xlabel('Bins')
plt.title('Unfolded vs Theory ' + varname)
plt.legend()
plt.show()

plt.plot(np.divide(unfolded, theory_xsec), '-o')
plt.ylim(0.95, 1.05)
plt.ylabel('Data/Theory')
plt.xlabel('Bins')
plt.title('Ratio ' + varname)
plt.show()
```
