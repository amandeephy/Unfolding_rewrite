```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
```

```python
fileptr_old = uproot.open("/depot/cms/top/bakshi3/TopSpinCorr_Run2_alt_version/CMSSW_10_6_29/src/TopAnalysis/Configuration/analysis/diLeptonic/UnfoldingHistos_2016preVFP/Nominal/emu/histosTUnfold_emu_ttbarsignalviatau_fromDilepton_2016ULpreVFP.root")
fileptr_new = uproot.open("/depot/cms/top/bakshi3/TopSpinCorr_Run2_generalized_ND/CMSSW_10_6_30/src/TopAnalysis/Configuration/analysis/diLeptonic/UnfoldingHistos_2016preVFP/Nominal/emu/histosTUnfold_emu_ttbarsignalplustau_fromDilepton_2016ULpreVFP.root")
```

```python
common_keys = [k for k in fileptr_old.keys() if 'reco' in k]
```

```python
bins = np.linspace(1,48,48)

for ck in common_keys :
    if 'pt' in ck or 'mass' in ck or 'P' in ck or 'M' in ck or 'ttbar' in ck : continue
        
    dist_1D_old = fileptr_old[ck].to_numpy()
    dist_1D_new = fileptr_new[ck].to_numpy()
    
    if len(dist_1D_old) > 2 : continue
        
    ratio = np.divide(dist_1D_new[0], dist_1D_old[0])
    
    _, ax = plt.subplots(figsize =(8, 6))
    ax.scatter(bins, ratio)
    ax.set_xlabel(ck)
#     print('Original yield :: ' + str(sum(dist_1D_old[0])))
#     print('New yield :: '      + str(sum(dist_1D_new[0])))
    print('Ratio yields  :: ' +  str(sum(dist_1D_old[0]) / sum(dist_1D_new[0])) )
```
