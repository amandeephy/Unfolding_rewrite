```python
# Here the idea should be to rebin 24x48 -> 12x24
# bins (2i, 2i+1) cartesian bins (2j, 2j+1) -> bin (i,j)

Resp_mat_1D_rebinned_to_12_24 = np.zeros((12,24))

for i in range(12) :
    for j in range(24) :
        Resp_mat_1D_rebinned_to_12_24[i][j] += Resp_mat_1D[2*i][2*j]   + Resp_mat_1D[2*i][2*j+1]
        Resp_mat_1D_rebinned_to_12_24[i][j] += Resp_mat_1D[2*i+1][2*j] + Resp_mat_1D[2*i+1][2*j+1]
```

```python
plt.imshow(Resp_mat_1D_rebinned_to_12_24, norm=matplotlib.colors.LogNorm(), origin='lower')
plt.title('1D response matrix rebinned to 12x24')
```
